```python
# Insertar estudiante
def insertar_estudiante():
    nombre = input("Ingrese el nombre: ")
    edad = int(input("Ingrese la edad: "))
    carrera = input("Ingrese la carrera: ")

    estudiante = {"nombre": nombre, "edad": edad, "carrera": carrera}
    estudiantes.insert_one(estudiante)
    print(f"Estudiante {nombre} insertado.\n")

# Listar estudiantes
def listar_estudiantes():
    print("\nListado de estudiantes:")
    for est in estudiantes.find():
        print(f"Nombre: {est['nombre']}, Edad: {est['edad']}, Carrera: {est['carrera']}")
    print()

# Actualizar edad
def actualizar_edad():
    nombre = input("Ingrese el nombre del estudiante a actualizar: ")
    nueva_edad = int(input("Ingrese la nueva edad: "))

    resultado = estudiantes.update_one(
        {"nombre": nombre},
        {"$set": {"edad": nueva_edad}}
    )

    if resultado.matched_count > 0:
        print(f"Edad de {nombre} actualizada.\n")
    else:
        print("Estudiante no encontrado.\n")

# Eliminar estudiante
def eliminar_estudiante():
    nombre = input("Ingrese el nombre del estudiante a eliminar: ")

    resultado = estudiantes.delete_one({"nombre": nombre})

    if resultado.deleted_count > 0:
        print(f"Estudiante {nombre} eliminado.\n")
    else:
        print("Estudiante no encontrado.\n")

# Menú principal
def menu():
    while True:
        print("----- MENÚ DE ESTUDIANTES -----")
        print("1. Insertar estudiante")
        print("2. Listar estudiantes")
        print("3. Actualizar edad")
        print("4. Eliminar estudiante")
        print("5. Salir")

        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            insertar_estudiante()
        elif opcion == "2":
            listar_estudiantes()
        elif opcion == "3":
            actualizar_edad()
        elif opcion == "4":
            eliminar_estudiante()
        elif opcion == "5":
            print("Saliendo del programa...")
            break
        else:
            print("Opción inválida, intente de nuevo.\n")

# Ejecutar programa
menu()
```
